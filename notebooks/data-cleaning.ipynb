```python
import pandas as pd
```

```python
df=pd.read_excel(r"C:\Users\Dr Kulbir Singh\retail-business-analytics\data\Online Retail.xlsx")
```

```python
df
```

```python
# Remove cancelled orders
df[df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
# Remove negative or zero quantities
df[df['Quantity'] <= 0]
```

```python
df = df[df['Quantity'] > 0]
```

```python
# Remove zero or negative prices
df[df['UnitPrice'] <= 0]
```

```python
df = df[df['UnitPrice'] > 0]
```

```python
# Handle missing customers
df.isna().sum()
```

```python
df = df.dropna(subset=['CustomerID'])
```

```python
df.isna().sum()
```

```python
df['CustomerID'] = df['CustomerID'].astype(int)
```

```python
# Datetime Features
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

df['order_date'] = df['InvoiceDate'].dt.date
df['year'] = df['InvoiceDate'].dt.year
df['month'] = df['InvoiceDate'].dt.month
df['month_year'] = df['InvoiceDate'].dt.to_period('M')
```

```python
df
```

```python
# Revenue & Order Metrics
df['revenue'] = df['Quantity'] * df['UnitPrice']
```

```python
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df[['Quantity','UnitPrice','revenue']].describe()
```

```python
# Create SQL-Ready Tables
customers = df[['CustomerID','Country']].drop_duplicates()
```

```python
customers
```

```python
products = df[['StockCode','Description','UnitPrice']].drop_duplicates()
```

```python
products
```

```python
orders = df[['InvoiceNo','CustomerID','order_date','Country']].drop_duplicates()
```

```python
order_items = df[['InvoiceNo','StockCode','Quantity','UnitPrice','revenue']]
```
